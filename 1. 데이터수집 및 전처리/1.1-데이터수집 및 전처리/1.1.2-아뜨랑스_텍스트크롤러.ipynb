```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import sys
import random
import os
import re
import math
import numpy as np
import pandas as pd
import xlwt

import cv2
import pyautogui
import imutils
import urllib
import datetime
```

```python
keyword_all = ['원피스', '스커트', '자켓', '니트',  '야상', '점퍼','패딩', '블라우스', '코트', '반팔 티셔츠', '맨투맨', '가디건', '청바지','스키니',  '셔츠', '치마','스웨터','후드티','면바지','캐주얼 바지','남방','트레이닝']
```

```python
path = 'c:/temp/chromedriver.exe'
driver = webdriver.Chrome(path)

url = "https://attrangs.co.kr/"
driver.get(url)
driver.maximize_window()
time.sleep(2)


# 팝업창 존재 시
try:
    driver.find_element_by_xpath('//*[@id="modal-popup"]/div/div[2]/a[2]').click()
except:
    print("팝업이 없습니다.")
```

```python
# 입력창에 키워드를 입력
def input_keyword():
    element = driver.find_element_by_id('keyword')
    element.send_keys(keyword)
    driver.find_element_by_xpath('//*[@id="header"]/div[1]/div/div/form/input[5]').click()

def doScrollDown_second(whileSeconds):
    start = datetime.datetime.now()
    end = start + datetime.timedelta(seconds=whileSeconds)
    while True:
        driver.execute_script('window.scrollTo(0, 1000);')
        time.sleep(1)
        if datetime.datetime.now() > end:
            break

# 자동 스크롤다운 함수
def scroll_down(driver):
    # driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    driver.execute_script("window.scrollTo(0, 5000)")

def scroll_up(driver):
    driver.execute_script("window.scrollTo(0,-6300);")
    time.sleep(5)


def scroll_down_oC(driver) :
    match_category_1 = match_center_loc(screenshot_1, img_down)
    pyautogui.moveTo(match_category_1)
    pyautogui.click()
    # time.sleep(2)

def match_center_loc(screen, template):
    res = cv2.matchTemplate(screen, template, cv2.TM_CCOEFF_NORMED)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    top_left = max_loc
    h, w = template.shape[:2]
    x, y = int(top_left[0] + w/2), int(top_left[1] + h/2)
    return x,y

def cnt_pa(driver):
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser').find('div','paging')
    a = soup.find('ol').find_all('li')

    return len(a)

# 사진 url 생성
def all_html(driver):
    time.sleep(1)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    dress_all = soup.find('ul', 'prd_basic col4').find_all('a')

    url_link = []
    for dress_img in dress_all:
            try:
                url_link.append(dress_img['href'])
            except:
                continue

    # url 중복 제거
    url_link = list(set(url_link))

    full_url = []
    for url_piece in url_link:
        url_full = 'https://attrangs.co.kr' + url_piece
        full_url.append(url_full)

    return full_url
```

```python
for keyword in keyword_all:
    input_keyword()
    time.sleep(1)
    cnt_all = cnt_pa(driver) # 전체 페이지 길이
    # debug
    print(cnt_all)

    urls = []
    for cnt in range(2, (cnt_all+1)):
        # doScrollDown_second(5)
        scroll_down(driver)
        full_url = all_html(driver) # url 파싱
        time.sleep(1)
        temp_urls = (full_url)

        cnt_count = cnt
        driver.find_element_by_xpath(f'//*[@id="big_section"]/div[3]/div/ol/li[{cnt_count}]').click() # 각 전체 페이지 클릭
        time.sleep(1)

        # url 예외 처리
        temp_urls = [item for item in temp_urls if item != 'https://attrangs.co.krjavascript:']
        urls.extend(temp_urls)

    print(str(len(urls)) + '개의 url을 가져옵니다.')
    # 각 url 들어가기
    content_list = []
    category_list = []
    for c in range(0, len(urls)):
        url = urls[c]
        try:
            driver.get(url)
            # time.sleep(1)
        except:
            print(url)
        
        # doScrollDown_second(4)
        scroll_down(driver)
        time.sleep(1)
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        # content 없을 시 다음 url로 넘어가기
        try:
            content_all = soup.find('div', 'fontBox').find_all('p')
        except:
            continue
        # time.sleep(1)

        # 한 게시글의 content를 한 string으로 묶어서 저장
        text = ''
        for content in content_all:
            text = text + (content.get_text()) + " "
        text.replace('\xa0', " ")   # 띄어쓰기가 \xa0으로 되어있는 것을 바꿔주기
        content_list.append(text)
        category_list.append(keyword)
        print(text)
        print (f'{c}번째 가져오기 완료')

    # time.sleep(1)

    url = "https://attrangs.co.kr/"
    driver.get(url)

    # 팝업창 존재 시
    try:
        driver.find_element_by_xpath('//*[@id="modal-popup"]/div/div[2]/a[2]').click()
    except:
        print("팝업이 없습니다.")
```

```python
import pandas as pd
f_dir = "D:/bunjang/data"

data = {'keyword': category_list,'text': content_list}
df = pd.DataFrame(data)
df.to_csv(f_dir + '/' + 'attrangs.csv')
```
