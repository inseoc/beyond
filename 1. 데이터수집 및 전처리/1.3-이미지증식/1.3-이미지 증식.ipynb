```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
# 넘파이로 변환해 놓은 사진 데이터 읽어 들이기
photos = np.load('D:/bunjang/cnn_photos.npz')
x = photos['x']
img = x[5] # 사진 하나 선택하기
plt.figure(figsize=(10,10))
plt.imshow(img)
    
plt.show()        
```

```python
plt.figure(figsize=(16,16))
# 16개만 확인
for i in range(16):
    plt.subplot(4,4,i+1)
    # 회전 시키기 
    center = (16, 16) # 회전 중심
    angle = i # * 2    # 각도 조정
    scale = 1.0   # 확대 비율
    mtx = cv2.getRotationMatrix2D(center, angle, scale)
    img2 = cv2.warpAffine(img, mtx, (64, 64))
    # 화면에 출력하기
    plt.imshow(img2)
plt.show()    
```
