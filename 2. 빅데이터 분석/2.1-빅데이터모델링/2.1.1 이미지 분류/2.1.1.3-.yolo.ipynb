```python
%cd /
from glob import glob
# 경로 내 train_data 및 validation_data 읽어오기
img_list_train = glob('/content/dataset/train/images/*.jpg')
img_list_val = glob('/content/dataset/valid/images/*.jpg')
print(len(img_list_train))
print(len(img_list_val))
```

```python
# 이미지 데이터들의 파일 경로를 텍스트에 넣기
with open('/content/dataset/train.txt', 'w') as f:
  f.write('\n'.join(img_list_train) + '\n')

with open('/content/dataset/val.txt', 'w') as f:
  f.write('\n'.join(img_list_val) + '\n')
```

```python
import yaml

with open('/content/dataset/data.yaml', 'r') as f:
  data = yaml.load(f)

print(data)

data['train'] = '/content/dataset/train.txt'
data['val'] = '/content/dataset/val.txt'

with open('/content/dataset/data.yaml', 'w') as f:
  yaml.dump(data, f)

print(data)
```
